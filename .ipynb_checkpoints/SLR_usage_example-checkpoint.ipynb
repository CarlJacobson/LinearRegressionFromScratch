```python
import cjlearn as cj
import numpy as np
```

```python
line = cj.random_line_maker(slope = 5, intercept = 40, variance = 10, points = 200)
```

```python
line.plot()
```

```python
model = cj.models.slr()
```

```python
model.fit(line)
```

```python
model.plot()
```

```python
model.predict(np.array([500,202,203,204,205]))
```

```python
model.beta0_hat
```

```python
model.beta1_hat
```

```python
model.formula()
```
